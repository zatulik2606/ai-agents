# Горячие клавиши Jupyter (Cursor / VS Code)

**Command mode** — `Esc` · **Edit mode** — `Enter`

| Категория | Действие | macOS | Windows / Linux |
|-----------|----------|-------|-----------------|
| Запуск | Запустить ячейку и перейти ниже | `Shift` + `Enter` | `Shift` + `Enter` |
| Запуск | Запустить ячейку (остаться на ней) | `Cmd` + `Enter` | `Ctrl` + `Enter` |
| Запуск | Запустить ячейку и вставить новую ниже | `Option` + `Enter` | `Alt` + `Enter` |
| Запуск | Прервать выполнение | `I`, `I` (command mode) | `I`, `I` (command mode) |
| Запуск | Перезапустить kernel | `Cmd` + `Shift` + `P` → Restart Kernel | `Ctrl` + `Shift` + `P` → Restart Kernel |
| Ячейки | Вставить ячейку выше | `A` | `A` |
| Ячейки | Вставить ячейку ниже | `B` | `B` |
| Ячейки | Удалить ячейку | `D`, `D` | `D`, `D` |
| Ячейки | Вырезать / копировать / вставить | `X` / `C` / `V` | `X` / `C` / `V` |
| Ячейки | Отменить удаление | `Z` | `Z` |
| Ячейки | Объединить ячейки | `Shift` + `M` | `Shift` + `M` |
| Ячейки | Разделить ячейку по курсору | `Cmd` + `Shift` + `-` | `Ctrl` + `Shift` + `-` |
| Ячейки | Тип → код / Markdown | `Y` / `M` | `Y` / `M` |
| Ячейки | Перейти выше / ниже | `↑` / `↓` | `↑` / `↓` |
| Ячейки | Редактировать ячейку | `Enter` | `Enter` |
| Код | Закомментировать строку | `Cmd` + `/` | `Ctrl` + `/` |
| Код | Курсор на строке ниже | `Option` + `Cmd` + `↓` | `Ctrl` + `Alt` + `↓` |
| Код | Выделить следующее вхождение | `Cmd` + `D` | `Ctrl` + `D` |
| Код | Мультикурсор | `Option` + клик | `Alt` + клик |
| Код | Перейти к определению | `F12` | `F12` |
| Код | Автодополнение | `Ctrl` + `Space` | `Ctrl` + `Space` |
| Прочее | Command mode | `Esc` | `Esc` |
| Прочее | Палитра команд | `Cmd` + `Shift` + `P` | `Ctrl` + `Shift` + `P` |
| Прочее | Установить пакет | `%pip install <пакет>` | `%pip install <пакет>` |

> Ячейки: command mode (`Esc`). Код: edit mode (`Enter`).

```python
%pip install numpy pandas matplotlib seaborn scikit-learn
```
